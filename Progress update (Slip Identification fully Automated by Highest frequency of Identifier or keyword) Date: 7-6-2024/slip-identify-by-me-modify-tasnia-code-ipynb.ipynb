```python
import pandas as pd

# Load the Excel file
file_path = '/kaggle/input/ocr500-transliterated/Ocr500_transliterated4.xlsx'
excel_data = pd.ExcelFile(file_path)

# Load the sheets into DataFrames
sheet1 = pd.read_excel(excel_data, sheet_name='Sheet1')
sheet3 = pd.read_excel(excel_data, sheet_name='Sheet3')

# Sort 'Sheet3' by 'Count of OCR_Text_Manual' in descending order
sorted_sheet3 = sheet3.sort_values(by='Count of OCR_Text_Manual', ascending=False)

# Extract sorted Row Labels
sorted_row_labels = sorted_sheet3['Row Labels'].tolist()

# Define a function to search for Row Labels in 'Extracted Text'
def search_row_labels_in_extracted_text(extracted_text, labels):
    for label in labels:
        if label in extracted_text:
            return label
    return None

# Apply the search function on 'Sheet1'
sheet1['Matched Label'] = sheet1['Extracted Text'].apply(lambda x: search_row_labels_in_extracted_text(x, sorted_row_labels))

# Save the updated DataFrame with the results
output_file_path = '/kaggle/working/path_to_your_updated_file_with_matched_labels.xlsx'
with pd.ExcelWriter(output_file_path) as writer:
    sheet1.to_excel(writer, sheet_name='Sheet1', index=False)
    sheet3.to_excel(writer, sheet_name='Sheet3', index=False)

print(f"Updated file saved at {output_file_path}")
```
